## NumPy Power Tools and optimization
NumPy is not just about storing data in arrays—it’s about
doing it efficiently. Beyond basicarray operations, these “power tools” help you manipulate, compute, and optimize withoutwasting time or memory.

Imagine working with a dataset of 10 million records. Using plain Python loops to process thesenumbers would take hours. NumPy power tools turn these operations into fast, vectorized, andmemory-efficient computations.

### Concept A: Fancy Indexing
Fancy indexing allows you to select elements of a NumPy array using
arrays of indices, insteadof plain slices. This is extremely useful when you need non-contiguous or complex selection.

```python
# Fancy indexing
import numpy as np

arr = np.array([10,20,30,40,50])
indices = [0,2,4]
selected = arr[indices]

print(selected) # Output: [103 0 50]
```

### Key points
* Can use integer arrays or boolean masks.
* Does not create a view; it returns a copy.
* Supports multi-dimensional arrays for row/column selection.

> Common misconception: Fancy indexing is not a slice. Changing 'selected' does not affect theoriginal array.

## Concept B: Universal Functions (ufuncs)
ufuncs are NumPy functions that operate element-wise on arrays. They are vectorized ,meaning they execute much faster than Python loops.
Examples of ufuncs: np.add, np.sqrt, np.exp, np.sin .

```python
# ufuncs
a = np.arange(4,14,2)
sqrt = np.sqrt(a)
print(a) # Output: [4 6 8 10 12]
```

```python
# ufuncs with broadcasting

arr = np.array([1, 2, 3])
print(np.add(arr,10)) # adds 10 to all the elements in the above array
```

### Key points:
* Support broadcasting automatically.
* Can perform operations on multiple arrays: np.add([1,2], [3,4])
* Support optional parameters like out for memory-efficient computations.

## Concept C: Memory Views
NumPy arrays support views—different arrays can share the same memory without copyingdata. Understanding views helps avoid unnecessary duplication.

```python
# Memory views
arr_2 = np.array([1,2,3,4])
view = arr[1:3]
view[0] = 100

print(f'Original array: {arr_2}')
print(f'View generated from original array: {view}') # Original array remains unchanged
```

## Concept D: Profiling
Profiling is the process of measuring code performance. NumPy offers tools to check executiontime and memory usage, helping optimize heavy computations.

```python
# profiling
import time
arr_3 = np.arange(1e7)

start = time.time() # start time
np.sqrt(arr_3) # program for which we want to check execution time
end = time.time()

print(f'Time: {round(end-start, 3)} ms')
```

### Key points:
* Compare vectorized NumPy code vs Python loops for performance.
* Profiling identifies bottlenecks before scaling.
* Tools like %timeit in Jupyter help benchmark functions easily.

### Fancy indexing in practice
Scenario:
You have exam scores for 100 students and want to select only the top 5 scores.

```python
# using argsort with fancy indexing to extract specific marks.
scores = np.array([55, 78, 92, 61, 85])

# np.argsort(scores) returns the indices that would sort the array.
top_indices = np.argsort(scores)[-3:] # last 3 for top scorers.
top_scores = scores[top_indices]

print(top_scores) # Output: [78 85 92]
```

### ufuncs for Bulk operations
Scenario:
Apply logarithm to all values in a large dataset to normalize it.

```python
# ufuncs usage with np.log
# np.log applies the logarithm to every element in the array at once
data = np.array([1, 10, 100, 1000])
log_data = np.log(data)

print(log_data) # Output: [0. 2.30258509 4.60517019 6.90775528]
```

### Memory views
Scenario:
You want to manipulate part of a large dataset without duplicating memory.

```python
arr = np.arange(10)
view = arr[2:5]
view[:] = 99
print(arr) # Output: [0 1 99 99 99 5 6 7 8 9]
```

### Explanation:
* Slicing
arr[2:5]
returns a view, sharing the same memory as the original array.
* Modifying
view
directly updates
arr
because both share memory.

### Profiling NumPy code
Scenario:
Compare performance of a Python loop vs NumPy vectorized operation.

```python
# using %timeit to benchmark performance
arr = np.arange(1e6)

# Vectorized
%timeit np.sqrt(arr)

# Loop
%timeit [x**0.5 for x in arr]
```

### Key Takeaway
Profiling reveals performance bottlenecks, demonstrating the efficiency advantage of vectorized NumPy operations over Python loops.

## Combining power tools

```python
data = np.random.randint(1, 100, 20)

# Fancy indexing
mask = data > 50
selected = data[mask]

# ufunc operation
sqrt_values = np.sqrt(selected)

# Memory view
view = sqrt_values[:5]
print("Original Data:", data)
print("Selected Data:", selected)
print("Square Roots (view):", view)
```
